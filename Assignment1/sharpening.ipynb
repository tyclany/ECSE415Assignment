1.2.1 Box blur and gaussian blur

```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt

# load image
image_rice_path = "rice.png"
img_rice = cv2.imread(image_rice_path)
# create a kernel for box filter
kernel = np.ones((5,5), np.float32)/(5*5)
# apply box filter
res1 = cv2.filter2D(img_rice, -1, kernel)

# get the sharpened image using formula
img_rice1 = (img_rice+(img_rice-res1))

# display image
plt.imshow(cv2.imread(image_rice_path))
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img_rice1)
plt.title("sharpened"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# load image
image_rice_path = "rice.png"
img_rice = cv2.imread(image_rice_path)

# apply gaussian filter
res2 = cv2.GaussianBlur(img_rice, (5,5), 2**(.5),2**(.5))

# get the sharpened image using formula
img_rice2 = img_rice+img_rice-res2

# display image
plt.imshow(cv2.imread(image_rice_path))
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img_rice2)
plt.title("sharpened"), plt.xticks([]), plt.yticks([])
plt.show()
```

1.2.2 The one using the gaussian filter should give a better result, since the gaussian filter gives a more smooth result.
