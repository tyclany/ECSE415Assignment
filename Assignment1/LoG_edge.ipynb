2.2 Laplacian of Gaussian

```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt

# load image with grayscale
image_woman_path = "woman.png"
img = cv2.imread(image_woman_path,cv2.IMREAD_GRAYSCALE)

# remove the blur first and then apply the laplacian of gaussian filter
res1 = cv2.GaussianBlur(img, (5,5), 0)
I_lap = cv2.Laplacian(res1,cv2.CV_32F, ksize=5)

# display the image
plt.imshow(I_lap, cmap="gray")
plt.title("Image"), plt.xticks([]), plt.yticks([])
plt.show()
```
