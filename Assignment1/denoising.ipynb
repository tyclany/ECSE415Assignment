1.1.1 7x7 gaussian

```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import compare_psnr
# load image
img_path = "gaussian_noise_peppers.png"
img_clean_path= "peppers.png"
img = cv2.imread(img_path)
img_clean = cv2.imread(img_clean_path)

# apply filter
res1 = cv2.GaussianBlur(img, (7,7), 2**(.5),2**(.5))

# covert to RGB 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)
img_clean = cv2.cvtColor(img_clean, cv2.COLOR_BGR2RGB)

# display image
plt.imshow(img_clean)
plt.title("Clean"), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(res1)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
compare_psnr(img_clean, res1)
```

1.1.2 box filter:

```python
image_path = "gaussian_noise_peppers.png"
img_clean_path= "peppers.png"
# load image
img = cv2.imread(image_path)
img_clean = cv2.imread(img_clean_path)

# apply filter
kernel = np.ones((7,7), np.float32)/(7*7)
#print("kernel:", kernel)
res2 = cv2.filter2D(img, -1, kernel)

# covert to RGB 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res2 = cv2.cvtColor(res2, cv2.COLOR_BGR2RGB)
img_clean = cv2.cvtColor(img_clean, cv2.COLOR_BGR2RGB)

# display image
plt.imshow(img_clean)
plt.title("Clean"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(res2)
plt.title("Box filter"), plt.xticks([]), plt.yticks([])
plt.show()
compare_psnr(img_clean, res2)
```

1.1.3 Compare psnr value of gaussiam and box filter: 29.87927167099926 and 28.278396476904042. Gaussian filter is better.

1.1.4 Salt and pepper gaussion filter

```python
# load image
img_path = "saltandpepper_noise_peppers.png"
img_clean_path= "peppers.png"
img = cv2.imread(img_path)
img_clean = cv2.imread(img_clean_path)

# apply filter
res1 = cv2.GaussianBlur(img, (7,7),2**(.5),2**(.5))

# convert to RGB for display
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)
img_clean = cv2.cvtColor(img_clean, cv2.COLOR_BGR2RGB)

# display image
plt.imshow(img_clean)
plt.title("Clean"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(res1)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
compare_psnr(img_clean, res1)
```

1.1.5 salt and pepper median filter

```python
# load image
image_path = "saltandpepper_noise_peppers.png"
img_path2= "peppers.png"
img = cv2.imread(image_path)
img2 = cv2.imread(img_path2)

# apply filter
res = cv2.medianBlur(img, 7)

# convert to RGB for display
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# display image
plt.imshow(img2)
plt.title("Clean"), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(res)
plt.title("Median"), plt.xticks([]), plt.yticks([])
plt.show()
# compare psnr
compare_psnr(img2, res)
```

1.1.6 Compare psnr value of guassian filter and median filter: 28.1671611567447 and 31.871009920638333. Median filter is better.
